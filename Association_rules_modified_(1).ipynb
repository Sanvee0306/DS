```python
# Import Data from CSV file 
import numpy as np
import pandas as pd 
dataset = pd.read_csv("groceries - groceries.csv",sep=',')
dataset = dataset.drop(['Item(s)'], axis=1)
dataset = dataset.replace(np.nan, '', regex=True)
dataset 
```

```python
dataset
transactions = []
for i in range(0, 9835):
    transactions.append([str(dataset.values[i,u]) for u in range(1, 31)]) 
```

```python
groceries_series  = pd.DataFrame(pd.Series(transactions))
groceries_series 
groceries_series.columns = ["business"]
groceries_series
```

```python
X = groceries_series['business'].str.join(sep=',').str.get_dummies(sep=',')
X
```

```python
!pip install mlxtend
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
frequent_itemsets = apriori(X, min_support=0.005,max_len=3,use_colnames = True)
frequent_itemsets 
```

```python
frequent_itemsets.sort_values('support',ascending = False,inplace=True)
frequent_itemsets.sort_values 
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head(20)
rules.sort_values('lift',ascending = False).head(10)  
```
